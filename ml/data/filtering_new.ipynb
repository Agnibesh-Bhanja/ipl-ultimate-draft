```python
# Weighted batting score (0-1)
df2024["BattingScore"] = (
    0.35 * df2024["Runs_norm"] +
    0.23 * df2024["BattingAverage_norm"] +
    0.24 * df2024["BattingStrikeRate_norm"] +
    0.07 * df2024["50s_norm"] +
    0.02 * df2024["100s_norm"] +
    0.03 * df2024["4s_norm"] +
    0.06 * df2024["6s_norm"]
)
# Convert to FIFA-style rating (50-99)
min_score = df2024["BattingScore"].min()
max_score = df2024["BattingScore"].max()

df2024["BattingRating"] = (
    65 +
    (df2024["BattingScore"] - min_score) /
    (max_score - min_score) * 34
).round().astype(int)
```

```python
def assign_role(row):
    if row["BattingRating"] >= 73 and row["BowlingRating"] >= 77:
        return "All-Rounder"
    elif row["BattingRating"] >= row["BowlingRating"]:
        return "Batsman"
    else:
        return "Bowler"
    
df2024["Role"] = df2024.apply(assign_role, axis=1)
```

```python
df2024["Role"].value_counts()
```

```python
def calculate_ovr(row):
    if row["Role"] == "Batsman":
        return round(
            0.90 * row["BattingRating"] +
            0.10 * row["BowlingRating"]
        )

    elif row["Role"] == "Bowler":
        return round(
            0.90 * row["BowlingRating"] +
            0.10 * row["BattingRating"]
        )

    elif row["Role"] == "All-Rounder":
        return round(
            0.55 * row["BattingRating"] +
            0.45 * row["BowlingRating"]
        )

    return round((row["BattingRating"] + row["BowlingRating"]) / 2)
```

```python
from collections import Counter

def get_team_counts(team_ids, cards):
    team = cards[cards["CardID"].isin(team_ids)]

    return {
        "Team": Counter(team["Team"]),
        "Role": Counter(team["Role"]),
        "CardType": Counter(team["CardType"])
    }


def player_chemistry(player, counts):

    chem = 0

    # ---------- Team ----------
    team_count = counts["Team"][player["Team"]] + 1

    if team_count >= 3:
        chem += 2
    elif team_count == 2:
        chem += 1

    # ---------- Role ----------
    role_count = counts["Role"][player["Role"]] + 1

    if player["Role"] == "Batsman" and role_count <= 5:
        chem += 1

    elif player["Role"] == "Bowler" and role_count <= 4:
        chem += 1

    elif player["Role"] == "All-Rounder" and role_count <= 2:
        chem += 1

    # ---------- Card Type ----------
    card_count = counts["CardType"][player["CardType"]] + 1

    if card_count >= 2:
        chem += 1

    return min(chem, 3)
```

```python
import random
import pandas as pd

def recommend_players(team_ids, cards, n=5):

    available = cards[~cards["CardID"].isin(team_ids)].copy()

    # Calculate counts ONLY ONCE
    counts = get_team_counts(team_ids, cards)

    recommendations = []

    for _, player in available.iterrows():

        chem = player_chemistry(player, counts)

        score = (
            player["OVR"] * 0.70 +
            chem * 8 +
            random.uniform(0, 5)
        )

        recommendations.append({
            "CardID": player["CardID"],
            "Name": player["Name"],
            "Team": player["Team"],
            "Role": player["Role"],
            "Year": player["Year"],
            "CardType": player["CardType"],
            "OVR": player["OVR"],
            "Chemistry": chem,
            "RecommendationScore": score
        })

    recommendations = pd.DataFrame(recommendations)

    recommendations = recommendations.sort_values(
        "RecommendationScore",
        ascending=False
    ).reset_index(drop=True)

    # -------- Tiers --------
    tier1 = recommendations.iloc[:5]
    tier2 = recommendations.iloc[5:15]
    tier3 = recommendations.iloc[15:40]

    # Random draft pattern
    patterns = [
        (2,2,1),
        (1,3,1),
        (1,2,2)
    ]

    weights = [0.7,0.2,0.1]

    elite, great, good = random.choices(
        patterns,
        weights=weights,
        k=1
    )[0]

    selected = pd.concat([
        tier1.sample(min(elite,len(tier1))),
        tier2.sample(min(great,len(tier2))),
        tier3.sample(min(good,len(tier3)))
    ])

    selected = selected.drop_duplicates("CardID")

    selected = selected.sample(frac=1).reset_index(drop=True)

    return selected
```

```python
draft_team = []

while len(draft_team) < 11:

    print(f"\nPick {len(draft_team)+1}")

    options = recommend_players(draft_team, cards)

    print(options[[
        "CardID",
        "Name",
        "Team",
        "Role",
        "OVR",
        "CardType"
    ]])

    choice = options.iloc[0]["CardID"]   # Auto-select first player
    print(f"Selected: {choice}")

    if choice not in draft_team:
        draft_team.append(choice)

print("Draft Complete!")
```
