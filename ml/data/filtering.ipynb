```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("ipl_ds2.csv")
df2=pd.read_csv("ipl_images.csv")
```

```python
df1.rename(columns={"player":"Name"},inplace=True)
```

```python
print(df1.shape)
print(df2.shape)
```

```python
df2.isnull().sum()
```

```python
df1["Name"] = df1["Name"].str.strip()
df2["Name"] = df2["Name"].str.strip()
```

```python
common_players = set(df1["Name"]).intersection(set(df2["Name"]))

print("Common Players:", len(common_players))
```

```python
only_in_df1 = set(df1["Name"]) - set(df2["Name"])
only_in_df2 = set(df2["Name"]) - set(df1["Name"])

print("Only in df1:", len(only_in_df1))
print("Only in df2:", len(only_in_df2))
```

```python
only_df1 = sorted(set(df1["Name"]) - set(df2["Name"]))
only_df2 = sorted(set(df2["Name"]) - set(df1["Name"]))

print("Only in df1:")
print(only_df1[:20])

print("\nOnly in df2:")
print(only_df2[:20])
```
